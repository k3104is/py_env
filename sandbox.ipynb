```python
import numpy as np
import matplotlib.pylab as plt
import random
import math
```

```python
def rulu(x):
    return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = rulu(x)
plt.plot(x, y)
plt.show()
```

```python
inner = 0
N = 100

figure, axes = plt.subplots()
draw_circle = plt.Circle((0, 0), 1.0,fill=False)

for _ in range(N):
    x1 = random.uniform(0, 1)
    y1 = random.uniform(0, 1)
    if (x1**2 + y1**2 <= 1):
        inner += 1
        draw_inner = plt.scatter(x1, y1 , c='red')
    else:
        draw_outer = plt.scatter(x1, y1, c='blue')
print(4 * inner / N)


axes.set_aspect(1)
axes.add_artist(draw_inner)
axes.add_artist(draw_outer)
axes.add_artist(draw_circle)



plt.grid(True)
plt.show()
```
